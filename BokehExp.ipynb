```
# Plot a complex chart with intearctive hover in a few lines of code

from bokeh.models import ColumnDataSource, HoverTool
from bokeh.plotting import figure
from bokeh.sampledata.autompg import autompg_clean as df
from bokeh.transform import factor_cmap

df.cyl = df.cyl.astype(str)
df.yr = df.yr.astype(str)

group = df.groupby(by=['cyl', 'mfr'])
source = ColumnDataSource(group)

p = figure(plot_width=800, plot_height=300, title="Mean MPG by # Cylinders and Manufacturer",
           x_range=group, toolbar_location=None, tools="")

p.xgrid.grid_line_color = None
p.xaxis.axis_label = "Manufacturer grouped by # Cylinders"
p.xaxis.major_label_orientation = 1.2

index_cmap = factor_cmap('cyl_mfr', palette=['#2b83ba', '#abdda4', '#ffffbf', '#fdae61', '#d7191c'], 
                         factors=sorted(df.cyl.unique()), end=1)

p.vbar(x='cyl_mfr', top='mpg_mean', width=1, source=source,
       line_color="white", fill_color=index_cmap, 
       hover_line_color="darkgrey", hover_fill_color=index_cmap)

p.add_tools(HoverTool(tooltips=[("MPG", "@mpg_mean"), ("Cyl, Mfr", "@cyl_mfr")]))

show(p)
```

```
import numpy as np
import pandas as pd
```

```
student_names = ["Student Name "+str(x) for x in range(1,30)]
year_list= np.random.randint(1,5,len(student_names))
noCompleted_list = np.random.randint(1,16,len(student_names))
total_score_list = noCompleted_list* np.random.randint(6, 12,len(student_names))

df_score = pd.DataFrame({
    "name":student_names
    ,"year":year_list
    ,"noCompleted":noCompleted_list
    ,"score":total_score_list
})
df_score.head()
```

```

df_score.info()
```

```
def plotStudents(df):
  
  #conver the dtypes for the group by as str 
  df['year'] = df.year.astype(str)
  #df['score']= df.score.apply(lambda x: str(x).rjust(3,"0"))
  #df['score'] = df.score.astype(str)
  df_score['score']= df_score.score.astype(int)
  
  
  #group students and year
  df = df.sort_values("noCompleted")
  dd= df.groupby(["year", "name"])
  #dd= df.groupby(["year", "name"]).agg('mean').reset_index()
  group= dd
  
  #define the source
  source = ColumnDataSource(group)
  
  #start plotting
  
  #define the figure outline
  p = figure(plot_width = 1200
            ,plot_height = 500
            ,title = "Yearwise Student Progress Tracker"
            ,x_range = group
            )
  
  p.xgrid.grid_line_color = None
  p.xaxis.major_label_orientation = 1.4
  
  index_cmap = factor_cmap('year_name', palette=['#2b83ba', '#2b88ba', '#ffffbf', '#fdae61', '#d7191c'], 
                         factors=sorted(df.year.unique()), end=1)
  
  #what has to be plotted
  p.vbar(x="year_name",width=1,top="noCompleted_mean", source=source, line_color="white",  fill_color=index_cmap)
  #p.vbar(x=["2"], width=1, top=[10], fill_color='red' )
  p.add_tools(HoverTool(tooltips=[("Score", "@noCompleted_mean")]))
  show(p)
  
plotStudents(df_score)
```

```
def plotStudents_1(df):
  
  #conver the dtypes for the group by as str 
  df['year'] = df.year.astype(str)
  #df['score']= df.score.apply(lambda x: str(x).rjust(3,"0"))
  #df['score'] = df.score.astype(str)
  df_score['score']= df_score.score.astype(int)
  
  
  #group students and year
  df = df.sort_values("noCompleted")
  #dd= df.groupby(["year", "name"])
  dd= df.groupby(["year", "name"]).agg('mean').reset_index()
  group= dd.groupby(["year", "name"])
  
  #define the source
  source = ColumnDataSource(group)
  
  #start plotting
  
  #define the figure outline
  p = figure(plot_width = 1200
            ,plot_height = 500
            ,title = "Yearwise Student Progress Tracker"
            ,x_range = group
            )
  
  p.xgrid.grid_line_color = None
  p.xaxis.major_label_orientation = 1.4
  
  index_cmap = factor_cmap('year_name', palette=['#2b83ba', '#2b88ba', '#ffffbf', '#fdae61', '#d7191c'], 
                         factors=sorted(df.year.unique()), end=1)
  
  #what has to be plotted
  p.vbar(x="year_name",width=1,top="noCompleted_mean", source=source, line_color="white",  fill_color=index_cmap)
  #p.vbar(x=["2"], width=1, top=[10], fill_color='red' )
  p.add_tools(HoverTool(tooltips=[("Score", "@noCompleted_mean")]))
  show(p)
  
plotStudents_1(df_score)
```

```
import datetime
completed_tuples=[]

for i in df_score.iterrows():
  s_date = datetime.datetime(2019, 8, np.random.randint(1,15), 23, 33, 31, 281368)
  days = 0
  for n in range(i[1]['noCompleted']):
    completed_tuples.append((i[1]['name'], "c"+str(np.random.randint(1,100)), s_date+datetime.timedelta(days=days)))
    days += 1

df_trend = pd.DataFrame(completed_tuples,columns=['name', 'node_name', 'time'])
df_trend.head()
```

```
df_trend.info()
```

```
df_trend_2 = df_trend.groupby([df_trend.time.dt.strftime("%d-%b"),'name'])['node_name'].agg("count").reset_index()
df_trend_3 = df_trend_2.groupby(['name','time']).sum().groupby(level=[0]).cumsum().reset_index()
df_trend_3[df_trend_3.name =="Student Name 1"]
```

```
def plotLines(df, students):
  x_range =[]
  for s in students:
    x_range.extend(df[df.name == s].time.unique().tolist())
  x_range = sorted(set(x_range))
  
  p = figure(x_range=x_range)
  for s in students:
    p.line(df[df.name == s].time.tolist(), df[df.name == s].node_name.tolist(), color='blue', line_width =2)
  
  show(p)
  
#plotLines(df_trend_3, ["Student Name 1", "Student Name 10","Student Name 18"])
plotLines(df_trend_3, df_trend_3.name.tolist())
```

```

import pandas as pd
import numpy as np
from bokeh.palettes import Spectral11
from bokeh.plotting import figure, show, output_file
output_file('temp.html')

toy_df = pd.DataFrame(data=np.random.rand(5,3), columns = ('a', 'b' ,'c'), index = pd.DatetimeIndex(start='01-01-2015',periods=5, freq='d'))   

numlines=len(toy_df.columns)
mypalette=Spectral11[0:numlines]

p = figure(width=500, height=300, x_axis_type="datetime") 
p.multi_line(xs=[toy_df.index.values]*numlines,
                ys=[toy_df[name].values for name in toy_df],
                line_color=mypalette,
                line_width=5)
show(p)
```
